```python
top_10_areas = df['Area'].value_counts().head(10)
top_10_areas
```

```python
#created a barchart

plt.figure(figsize=(10, 5))
plt.bar(top_10_areas.index,top_10_areas.values,color = "r")
plt.xlabel('Areas')
plt.ylabel('values')
plt.title('Areas with Most Restaurants')
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
        
```

```python
popular_food_types = df.groupby(['City', 'Food type'])['ID'].count().reset_index()
popular_food_types = popular_food_types.sort_values(by=['City', 'ID'], ascending=[True, False])
popular_food_types = popular_food_types.drop_duplicates(subset='City')
popular_food_types
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Basic barplot using seaborn
sns.barplot(data=popular_food_types, x='City', y='ID', hue='Food type')

# Titles and labels

plt.title("Popular Food Type Swiggy Served by Each City")
plt.xlabel("City")
plt.ylabel("No. of Restaurants")

# Rotate x labels
plt.xticks(rotation=45, ha='right')

# Show plot
plt.show()
```

```python
above_45_ratings = df[df['Avg ratings'] > 4.5]
below_45_ratings = df[df['Avg ratings'] <= 4.5]
percentage_above_45 = (len(above_45_ratings) / len(df)) * 100
percentage_below_45 = 100 - percentage_above_45

rating_df = pd.DataFrame({'Rating': ['Above 4.5', '4.5 and Below'],
                          'Percentage': [percentage_above_45, percentage_below_45]})
```

```python
above_4_rating = df[df['Avg ratings'] > 4]
below_4_rating = df[df['Avg ratings'] <= 4]

percentage_above_4 = (len(above_4_rating) / len(df)) * 100
percentage_below_4 = 100 - percentage_above_4

rating_df = pd.DataFrame({
    'rating': ['above 4', '4 and below'],
    'percentage': [percentage_above_4, percentage_below_4]
})

print(rating_df)


```

```python

# creating a pie chart - rating above 4 and below 4

c = ['yellow','red']

plt.pie(
    rating_df['percentage'],           
    labels=rating_df['rating'],        
    colors=c,
    autopct='%1.1f%%')


plt.title('rating percentage')
plt.show()
    


```
